## 4. Normalization Comparison

### 4.1 Compare different normalization methods

```python
print("Comparing normalization methods...")
normalization_results = {}
```

```python
# Store original state for comparison
original_stats = schic.describe_all()
```

```python
# Test each normalization method
for method in NORMALIZATION_METHODS:
    print(f"\n{'='*60}")
    print(f"Testing {method} normalization...")
    print(f"{'='*60}")
    
    # Create a config optimized for sparse data if needed
    if method == "SCALE":
        config = scHiCNormConfig.for_ultra_sparse_data()
    else:
        config = scHiCNormConfig(
            method=method,
            min_nnz=10,
            coverage_threshold=1,
            cis_only=True
        )
    
    # Run normalization
    result = schic.normalize(
        method=method,
        output_dir=OUTPUT_DIR / f"normalized_{method}",
        config=config,
        apply_to_data=False,  # Don't modify original data
        n_workers=4,
        show_progress=False  # Disable progress bars
    )
    
    # Store results
    normalization_results[method] = {
        'result': result,
        'success_rate': len(result.cool_paths) / len(schic.cell_names) * 100,
        'failed_cells': len(result.failed_cells)
    }
    
    print(f"{method}: {len(result.cool_paths)}/{len(schic.cell_names)} cells normalized")
    print(f"  Success rate: {normalization_results[method]['success_rate']:.1f}%")
```

### 4.2 Apply best performing normalization method

```python
# Find method with highest success rate
best_method, best_info = max(
    normalization_results.items(),
    key=lambda kv: kv[1]['success_rate']
)
best_rate = best_info['success_rate']

print(f"\nBest performing method: {best_method}")
print(f"Success rate: {best_rate:.1f}%")
```

```python
# Apply the best normalization
print(f"\nApplying {best_method} normalization to data...")
best_result = normalization_results[best_method]['result']

if best_result.weights:
    schic._apply_normalization_weights(best_result.weights)
    schic.is_normalized = True
    schic.normalization_method = best_method
    print("Normalization weights applied")
else:
    print("No weights available, normalization may have failed")
```

### 4.3 Compare statistics before and after normalization

```python
if schic.is_normalized:
    # Get normalized statistics
    normalized_stats = schic.describe_all()
    
    # Create comparison DataFrame
    comparison_data = {
        'Metric': [],
        'Before Normalization': [],
        'After Normalization': [],
        'Change (%)': []
    }
    
    # Compare key metrics
    metrics_to_compare = [
        'Average Mean Contact Value',
        'Average Std Contact Value',
        'Average Non-zero Contacts'
    ]
    
    for metric in metrics_to_compare:
        if metric in original_stats and metric in normalized_stats:
            before = original_stats[metric]
            after = normalized_stats[metric]
            change = ((after - before) / before * 100) if before != 0 else 0
            
            comparison_data['Metric'].append(metric)
            comparison_data['Before Normalization'].append(before)
            comparison_data['After Normalization'].append(after)
            comparison_data['Change (%)'].append(change)
    
    comparison_df = pd.DataFrame(comparison_data)
    print("\nNormalization Effect Summary:")
    display(comparison_df.round(4))
```

### Common issues and solutions:

1. **All cells fail normalization**: 
    - Use `scHiCNormConfig.for_ultra_sparse_data()`
    - Try `method='SCALE'` for simple scaling
    - Lower `coverage_threshold` and `min_nnz` parameters
______________________________________________________________
2. **Memory issues**:
    - Process in smaller batches using `chunk_size` parameter
    - Use `n_workers=1` for serial processing
    - Enable `use_float32=True` in config
______________________________________________________________
3. **Slow performance**:
    - Increase `n_workers` for parallel processing
    - Disable `diagnose_first=False` if you know your data
    - Use `show_progress=True` to monitor progress
